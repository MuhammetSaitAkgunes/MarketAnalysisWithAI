```python
import pandas as pd
import seaborn as sbn
import numpy as np
```

```python
dataFrame = pd.read_excel("C:/Users/Muhammet Sait/Desktop/deneme.xlsx")
```

```python
sbn.pairplot(dataFrame)
```

```python
from sklearn.model_selection import train_test_split
```

```python
satis_fiyat = dataFrame["dolar_satis"].values
degerler = dataFrame[["dolar_6aylik_faiz","dolar_12aylik_faiz","dolar_12+aylik_faiz","tufe_12aylik"]].values


fiyat_train, fiyat_test, deger_train, deger_test = train_test_split(satis_fiyat,degerler,test_size=0.33,random_state=15)
```

```python
deger_test.shape
```

```python
deger_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(degerler)
```

```python
deger_train = scaler.transform(deger_train)
deger_test = scaler.transform(deger_test)
```

```python
deger_train
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(7,activation="relu"))
model.add(Dense(7,activation="relu"))
model.add(Dense(7,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="rmsprop",loss="mse")
```

```python
model.fit(deger_train, fiyat_train, epochs=750)
```

```python
loss = model.history.history["loss"]
```

```python
sbn.lineplot(x=range(len(loss)),y = loss)
```

```python
trainLoss = model.evaluate(deger_train,fiyat_train, verbose=0)
```

```python
testLoss = model.evaluate(deger_test,fiyat_test, verbose=0)
```

```python
trainLoss
```

```python
testLoss
```

```python
testTahminleri = model.predict(deger_test)
```
